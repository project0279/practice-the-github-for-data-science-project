```python
import numpy as np
import pandas as pd

df_movies = pd.read_csv('tmdb_5000_movies.csv')
df_credits = pd.read_csv('tmdb_5000_credits.csv')

df_credits.head(1)
```

```python
# MERGING THE BOTH DATAFRAME
df_merge=df_movies.merge(df_credits,on='title')

# MAKE THE COPY
df_copy = df_merge.copy(deep=True)

# columns that we includes 
# 1 genres id keyword title overview cast crew

df_copy=df_copy[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# find the null values and drop them
df_copy.isnull().sum()

df_copy.dropna(inplace=True)

# check the duplicate values
df_copy.duplicated().sum()
```

```python
df_copy.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# now convert this into [Action,fantasy,scifi]
```

```python
import ast
def convert(obj):
    if isinstance(obj, str):
        hai= [i['name'] for i in ast.literal_eval(obj)]
        return hai
    elif isinstance(obj, list):
        return obj  # already converted
    else:
        return []
```

```python
# now apply this convert into genres
df_copy['genres'] = df_copy['genres'].apply(convert)
```

```python
df_copy['genres']
```

```python
df_copy['keywords'] = df_copy['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0

    # if obj is string → convert
    if isinstance(obj, str):
        lis = ast.literal_eval(obj)

    for i in lis:
        if counter < 3:
            # if dict → extract name
            if isinstance(i, dict):
                L.append(i['name'])
            else:
                L.append(i)
            counter += 1
        else:
            break

    return L
```

```python
df_copy['cast']=df_copy['cast'].apply(convert3)
```

```python
df_copy.head()
```

```python
# only get the director values
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L  
```

```python
df_copy['crew'] = df_copy['crew'].apply(director)
```

```python
df_copy.head(1)
```

```python
# convert the overview into list
df_copy['overview'] = df_copy['overview'].apply(lambda x:x.split())# it convert the string into list not in data_type
```

```python
df_copy['overview'].head()
```

```python
for col in ['crew','genres','keywords','cast']:
    df_copy[col] = df_copy[col].apply(lambda x: [i.replace(" ","") for i in x] if isinstance(x, list) else [])
```

```python
df_copy.head(1)
```

```python
#concat the columns
df_copy['tags'] =  df_copy['overview']+ df_copy['genres']+ df_copy['keywords']+ df_copy['cast']+ df_copy['crew']
new_df_copy = df_copy[['movie_id','title','tags']]
new_df_copy
```

```python
# convert it back  into string
new_df_copy['tags'] = new_df_copy['tags'].apply(lambda x:' '.join(x))
# join=is used so it combine multiple string in the list into single list
# ' ' = tells to make a gap b/w words 

```

```python
# 
new_df_copy.head(1)
```

```python
# string
new_df_copy['tags'][0]
```

```python
# recommended make lower case
new_df_copy['tags'] = new_df_copy['tags'].apply(lambda x:x.lower())
```

```python
new_df_copy['tags'][0]
```

```python
# It cuts off endings of words so that similar words are #    treated as the same word. connections → connect
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() # Creates a stemmer object that can reduce words to their root form.
# like running => run
def stem(text):
    y=[]
    for i in text.split(): 
        y.append(ps.stem(i)) # Splits the sentence into individual words in a list format
        
    return " ".join(y) # after convert list back to string

new_df_copy['tags']=new_df_copy['tags'].apply(stem)
     
```

```python
# now for similarity calculate b/w movies by vectorisation
#CountVectorizer = convert text->vectors
# so it count how many time a word come 
# and use cosine similarity means find the distance among 1 movies with all the movie in the vector plane 

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df_copy['tags']).toarray()

vector[0]# first movie Avatar
```

```python
# name of 5000 features name
cv.get_feature_names_out()
```

```python
# FIND THE DISTANCE AMONG VECOTRS BY USING ANGLE 
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity.shape
#it calculate the similarity of single vector by each and every vector
```

```python
# function for geting the index of the similar of the movie

def recommend(movie):
    movie_index=new_df_copy[new_df_copy['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
 
    for i in movies_list:
        print(new_df_copy.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
